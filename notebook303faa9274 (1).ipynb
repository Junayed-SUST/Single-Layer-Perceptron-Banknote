```python
import pandas as pd
import os


file_path = None

for root, _, files in os.walk("/kaggle/input"):
    for f in files:
        if f.endswith(".csv") or f.endswith(".txt"):
            if "bank" in f.lower():
                file_path = os.path.join(root, f)

if file_path is None:
    raise Exception("Dataset file not found in Kaggle input!")

print("Found Dataset:", file_path)


df = pd.read_csv(file_path)

print("\nPreview:")
print(df.head())


df.columns = df.columns.str.strip()
df = df.dropna().reset_index(drop=True)

print("\nColumns:", df.columns)



possible_labels = ["class", "Class", "Class_Label"]

label_col = None
for c in df.columns:
    if c in possible_labels:
        label_col = c

if label_col is None:
    label_col = df.columns[-1]  # fallback

# feature selection (first 3 numeric cols)
X = df.iloc[:, 0:3].values.tolist()
y = df[label_col].astype(int).tolist()

print("\nTotal Samples:", len(X))



class Perceptron:

    def __init__(self, n_inputs, lr=0.1):
        self.w = [0.0]*n_inputs
        self.b = 0.0
        self.lr = lr

    def step(self, z):
        return 1 if z >= 0 else 0

    def forward(self, x):
        z = self.b
        for i in range(len(x)):
            z += x[i]*self.w[i]
        return self.step(z)

    def train(self, X, y, epochs=50):
        for epoch in range(epochs):
            error = 0

            for xi, yi in zip(X, y):
                pred = self.forward(xi)
                err = yi - pred
                error += abs(err)

                for i in range(len(self.w)):
                    self.w[i] += self.lr * err * xi[i]

                self.b += self.lr * err

            print(f"Epoch {epoch+1} | Error: {error}")

            if error == 0:
                print("Converged!")
                break



model = Perceptron(3, 0.1)
model.train(X, y)


pred = [model.forward(x) for x in X]

acc = sum([1 for a,p in zip(y,pred) if a==p]) / len(y) * 100   

print("\nAccuracy:", acc)
```
